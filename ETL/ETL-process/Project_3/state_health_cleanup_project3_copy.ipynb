```python
import pandas as pd
from functools import reduce

path2009 = "../../Resources/2009-Annual.csv"
path2010 = "../../Resources/2010-Annual.csv"
path2011 = "../../Resources/2011-Annual.csv"
path2012 = "../../Resources/2012-Annual.csv"
path2013 = "../../Resources/2013-Annual.csv"
path2014 = "../../Resources/2014-Annual.csv"
path2015 = "../../Resources/2015-Annual.csv"
path2016 = "../../Resources/2016-Annual.csv"
path2017 = "../../Resources/2017-Annual.csv"
path2018 = "../../Resources/2018-Annual.csv"
path2019 = "../../Resources/2019-Annual.csv"

data2009 = pd.read_csv(path2009)
data2010 = pd.read_csv(path2010)
data2011 = pd.read_csv(path2011)
data2012 = pd.read_csv(path2012)
data2013 = pd.read_csv(path2013)
data2014 = pd.read_csv(path2014)
data2015 = pd.read_csv(path2015)
data2016 = pd.read_csv(path2016)
data2017 = pd.read_csv(path2017)
data2018 = pd.read_csv(path2018)
data2019 = pd.read_csv(path2019)

data2009.head()
```

```python
ap2009 = data2009.loc[(data2009['Measure Name'] == 'Air Pollution')]
ap2009 = ap2009[['Edition','State Name','Value']]
ap2009 = ap2009.rename(columns={'Value':'air_pollution_val'})
cancer2009 = data2009.loc[(data2009['Measure Name'] == 'Cancer Deaths')]
cancer2009 = cancer2009[['Edition','State Name','Value']]
cancer2009 = cancer2009.rename(columns={'Value':'cancer_death_val'})
cardio2009 = data2009.loc[(data2009['Measure Name'] == 'Cardiovascular Deaths')]
cardio2009 = cardio2009[['Edition','State Name','Value']]
cardio2009 = cardio2009.rename(columns={'Value':'cardio_death_val'})
childpov2009 = data2009.loc[(data2009['Measure Name'] == 'Children in Poverty')]
childpov2009 = childpov2009[['Edition','State Name','Value']]
childpov2009 = childpov2009.rename(columns={'Value':'child_pov_val'})
choles2009 = data2009.loc[(data2009['Measure Name'] == 'Cholesterol Check')]
choles2009 = choles2009[['Edition','State Name','Value']]
choles2009 = choles2009.rename(columns={'Value':'choles_check_val'})
dentvis2009 = data2009.loc[(data2009['Measure Name'] == 'Dental Visit, Annual')]
dentvis2009 = dentvis2009[['Edition','State Name','Value']]
dentvis2009 = dentvis2009.rename(columns={'Value':'dent_vis_val'})
dentnum2009 = data2009.loc[(data2009['Measure Name'] == 'Dentists')]
dentnum2009 = dentnum2009[['Edition','State Name','Value']]
dentnum2009 = dentnum2009.rename(columns={'Value':'dentists_val'})
diabetes2009 = data2009.loc[(data2009['Measure Name'] == 'Diabetes')]
diabetes2009 = diabetes2009[['Edition','State Name','Value']]
diabetes2009 = diabetes2009.rename(columns={'Value':'diabetes_val'})
drug2009 = data2009.loc[(data2009['Measure Name'] == 'Drug Deaths')]
drug2009 = drug2009[['Edition','State Name','Value']]
drug2009 = drug2009.rename(columns={'Value':'drug_deaths_val'})
healthstatfem2009 = data2009.loc[(data2009['Measure Name'] == 'Health Status - Female')]
healthstatfem2009 = healthstatfem2009[['Edition','State Name','Value']]
healthstatfem2009 = healthstatfem2009.rename(columns={'Value':'health_stat_fem_val'})
immunchild2009 = data2009.loc[(data2009['Measure Name'] == 'Immunizations - Children')]
immunchild2009 = immunchild2009[['Edition','State Name','Value']]
immunchild2009 = immunchild2009.rename(columns={'Value':'immun_child_val'})
incomeineq2009 = data2009.loc[(data2009['Measure Name'] == 'Income Inequality')]
incomeineq2009 = incomeineq2009[['Edition','State Name','Value']]
incomeineq2009 = incomeineq2009.rename(columns={'Value':'income_ineq_val'})
infantmort2009 = data2009.loc[(data2009['Measure Name'] == 'Infant Mortality')]
infantmort2009 = infantmort2009[['Edition','State Name','Value']]
infantmort2009 = infantmort2009.rename(columns={'Value':'infant_mort_val'})
infectdis2009 = data2009.loc[(data2009['Measure Name'] == 'Infectious Disease')]
infectdis2009 = infectdis2009[['Edition','State Name','Value']]
infectdis2009 = infectdis2009.rename(columns={'Value':'infect_dis_val'})
obesity2009 = data2009.loc[(data2009['Measure Name'] == 'Obesity')]
obesity2009 = obesity2009[['Edition','State Name','Value']]
obesity2009 = obesity2009.rename(columns={'Value':'obesity_val'})
physicalinac2009 = data2009.loc[(data2009['Measure Name'] == 'Physical Inactivity')]
physicalinac2009 = physicalinac2009[['Edition','State Name','Value']]
physicalinac2009 = physicalinac2009.rename(columns={'Value':'phys_inac_val'})
premdeath2009 = data2009.loc[(data2009['Measure Name'] == 'Premature Death')]
premdeath2009 = premdeath2009[['Edition','State Name','Value']]
premdeath2009 = premdeath2009.rename(columns={'Value':'prem_death_val'})
smoking2009 = data2009.loc[(data2009['Measure Name'] == 'Smoking')]
smoking2009 = smoking2009[['Edition','State Name','Value']]
smoking2009 = smoking2009.rename(columns={'Value':'smoking_val'})
uninsured2009 = data2009.loc[(data2009['Measure Name'] == 'Uninsured')]
uninsured2009 = uninsured2009[['Edition','State Name','Value']]
uninsured2009 = uninsured2009.rename(columns={'Value':'uninsured_val'})
uninsured2009.head()
```

```python
dfs_2009 = [ap2009, cancer2009, cardio2009, childpov2009, choles2009, dentvis2009, dentnum2009, diabetes2009, drug2009, healthstatfem2009, immunchild2009, incomeineq2009, infantmort2009, infectdis2009, obesity2009, physicalinac2009, premdeath2009, smoking2009, uninsured2009]
nan_value = 0
clean2009 = reduce(lambda  left,right: pd.merge(left,right,on=['Edition', 'State Name'],
                                           how='outer'), dfs_2009)

clean2009 = clean2009.drop_duplicates(subset=['State Name'])

clean2009
```
